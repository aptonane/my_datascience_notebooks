# EDA (Exploratory Data Analysis) dos Acidentes nas Rodovias Federais Brasileiras


A Polícia Rodoviária Federal – PRF – atende cerca de 70.000 km de rodovias federais e está distribuída em todo o território nacional.<br>
Entre os anos de 2007 e 2016 o registro de acidentes era realizado por meio do sistema onde o policial responsável pela ocorrência inseria os dados referentes aos envolvidos, ao local, aos veículos e ao acidente.
Em janeiro de 2017 este sistema (BR-Brasil) foi descontinuado e a PRF passou a utilizar um novo sistema para registro das ocorrências de acidentes de trânsito. 
Este novo sistema fornece uma descrição sucinta das variáveis presentes nos conjuntos de dados registrados a partir de janeiro de 2017, principalmente relativas às pessoas envolvidas.

Nos anos de 2015, 2016 e 2017 somados, ceca de 14,000 pessoas morreram nas estradas federais e 200,000 ficaram feridas. No total (mortos, feridos e ilesos) cerca de 510.000 foram vitimas de trânsito. 
Os números levantados neste EDA são alarmante e preocupantes.<br>

Observamos, também, que a causa principal dos acidentes registrada foi **falta de atenção**. Mas, pelos relatórios analisados em paralelo como da CNT (ver abaixo), a maior causa dos acidentes tem sido as condicoes precarias de conservação da malha viária brasileira. Talvez os erros do motorista quanto a falta de atenção (que deve ser redobrada quando a situação das estradas está inadequada), velocidade incompatível, ingestão de álcool, não guardar distancia de seguranca e desobediência à sinalização sejam agravados alidos à condição precária das estradas, seja a grande causa da quantidade excessiva de acidentes e mortes nas estradas brasileiras.

A primeira parte desta EDA, foi baseada no sistema anterior (já que foram analisados anos anteriores à reformulação dos sistema), focando nas variáveis disponíveis.
Paral tal, utilizei o banco de dados dos últimos 3 anos (2015, 2016, 2017. Esses bancos de dados podem ser encontrados no endereço: https://www.prf.gov.br/portal/dados-abertos/acidentes  - (bancos de dados utilizados: "datatran2017.csv", "datatran2016_atual.csv", "datatran2015.csv".

É importante destacar que os problemas envolvendo hoje as estradas brasileiras podem ser considerados agravantes dos acidentes que vem vitimando muitas pessoas nas estradas brasileiras. 
Abaixo apresento uma descriçao sucinta da situaçao atual das estradas brasileiras e que tem como fonte trabalhos elaborados pela instituição brasileira CNT (Confederação Nacional dos Transportes) - ver abaixo.<br>

Para se ter uma idéia, a pesquisa da CNT de Rodovias 2017  (ver abaixo) avaliou 105.814 km de rodovias e identificou que a malha pavimentada brasileira desde muito tempo continua a oferecer condições inadequadas aos seus usuários, de forma que 61,8 por cento da extensão pesquisada foi classificada como Regular, Ruim ou Péssimo no Estado Geral. Este trabalho ainda diz que para melhorar essas condições, as intervenções necessárias são mais complexas e possuem maior custo. Muitas rodovias brasileiras ainda mantêm o traçado original de quando foram projetadas, não levando em consideração o aumento do fluxo de veículos ao longo dos anos.

O relatório na CNT menciona que a característica de pior resultado foi a Geometria da Via (traçado da rodovia, existência de curvas perigosas, acostamentos, pontes, viadutos, faixas adicionais de subida etc), que teve avaliação Regular, Ruim ou Péssimo em 77,9 por cento da extensão pesquisada. Esse resultado é reflexo da falta de intervenções de adequação das rodovias brasileiras que, em sua maioria, foram construídas na década de 1970 e ainda são majoritariamente de pista simples (que requer ultrapassagem passando pela pista de sentido contrário), ainda que o fluxo por elas atendido atualmente exija uma maior capacidade de tráfego. A sinalização teve a mesma classificação desfavorável em 59,2 por cento da extensão, enquanto o Pavimento foi considerado inadequado em 50,0 por cento, ou seja, 52.911 km.<br>

Segundo esta mesma pesquisa, um outro problema está relacionado à estrutura dos pavimentos flexíveis, que não atendem  às exigências técnicas tanto da capacidade de suporte das camadas do pavimento como da qualidade dos materiais empregados no revestimento. Falhas construtivas têm como consequência um processo de deformação mais acelerado, resultando em maiores custos com a reparação desses pavimentos para atingir condições ideais de tráfego.<br>

O segundo trabalho da CNT (ver abaixo) diz que essa situação de inadequação da malha elevam os custos do transportador, reduzem a segurança das vias e comprometem uma retomada sustentada do crescimento econômico do país, dado que o transporte rodoviário é o principal meio de deslocamento de pessoas e cargas em território nacional.<br>

Este mesmo trabalho diz que mesmo nos anos de maior volume de investimentos em rodovias, o governo federal não desembolsou sequer 0,5 por cento do PIB em melhorias da infraestrutura. Desde 2004, os anos de 2010 e 2011 registraram o maior investimento federal na malha rodoviária: 0,26 por cento e 0,25 por cento do PIB, respectivamente. <br>
Além de ser insuficiente, o volume vem diminuindo, tendo chegado a 0,14 por cento do Produto Interno Bruto, em 2016.<br>
As deficiências nas rodovias conforme dados do relatório da CNT potencializam tanto a ocorrência de acidentes quanto a sua gravidade.

Ainda segundo o mesmo relatório da CNT,  em 2016, o governo federal brasileiro investiu 8,61 bilhões de reais (0.14 por cento do PIB) em intervenções nas rodovias públicas federais, sendo 64,3 por cento em ações de manutenção de rodovias distribuídas entre todas as regiões. Já em 2017, até junho, foram desembolsados
3,01 bilhões de reais, dos quais 65,6 por cento também foram alocados em manutenção. Apesar do aumento dos aportes em 2016, em relação ao ano de 2015 (5,95 bilhões de reais), a qualidade das rodovias piorou, comprometendo ainda mais a atividade econômica. <br> 

Assim, a análise conjunta dos resultados da Pesquisa CNT, deste trabalho e da execução orçamentária do governo federal aponta para uma gestão inadequada dos recursos públicos. Isto é, ainda que verbas públicas tenham sido destinadas para a correção dos problemas, as intervenções não foram eficazes no propósito de melhoria
da malha rodoviária federal e acidentes continuam acontecendo e matando  e ferindo pessoas. <br>

Também segundo os relatórios da CNT, seriam necessários cerca de 293,88 bilhões de reias (estimados pelo Plano CNT de Transporte e Logística) para solucionar os entraves existentes no sistema rodoviário brasileiro de forma eficiente.<br>

Os dados levantados por este trabalho que visam retratar e detalhar a situção atual dos acidentes ocorridos e do número de vítimas como forma de conscientização, informação e ação.<br>

# 1.13) Análise BR 101
Construída pelo Exército Brasileiro, atravessa doze estados brasileiros:
- Rio Grande do Norte
- Paraíba
- Pernambuco
- Alagoas
- Sergipe
- Bahia
- Espírito Santo
- Rio de Janeiro
- São Paulo
- Paraná
- Santa Catarina
- Rio Grande do Sul <br>
Em toda sua extensão é denominada oficialmente Rodovia Governador Mário Covas.
Segue no sentido norte-sul por praticamente todo o litoral leste brasileiro, do Rio Grande do Norte ao Rio Grande do Sul.<br> 
Inicialmente, possuía um trecho não construído entre o município paulista de Peruíbe e Garuva (Santa Catarina). No trecho incompleto entre Peruíbe (SP) e Garuva (SC), a BR-101 é sobreposta às rodovias BR-116 (Rodovia Régis Bittencourt) e BR-376. Esses trechos são computados na quilometragem total da rodovia pelo Ministério dos Transportes.<br>
**Por este motivo, na análise da Br 101, contabilizarei também os acidentes ocorridos nas Br's 116 e 376**
